```python
import os
import csv
import psycopg2

from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# File names
folder_path = os.getcwd()  # change if needed

file_names = [os.path.splitext(f)[0] for f in os.listdir(folder_path) if f.endswith(".csv")]

print (file_names)
```

```python
len(file_names)
```

```python
# Get the connection string from the environment variable
conn_string = os.getenv("DATABASE_URL")

try:
    with psycopg2.connect(conn_string) as conn:
        print("Connection established")
        
        with conn.cursor() as cur:
        
            for file in file_names:
                try:
                    # Open and read your CSV file
                    with open(f'{file}.csv', 'r') as f:
                        # Skip the header row if your CSV has one
                        next(f)

                        # Use COPY to efficiently load data
                        cur.copy_expert(
                            f"COPY {file} FROM STDIN WITH CSV",
                            f
                        )
                    conn.commit()
                    print(f"Data {file} uploaded successfully\n")
                except Exception as e:
                    conn.rollback()
                    print(f"Failed: {file}")
                    print(f"Error: {e}")

except Exception as e:
    print("Connection failed.")
    print(e)
```
